```python
my_color = LinearSegmentedColormap.from_list('my_gradient', (
    # Edit this gradient at https://eltos.github.io/gradient/#FF0000-FFFFFF-0077FC
    (0.000, (1.000, 0.000, 0.000)),
    (0.500, (1.000, 1.000, 1.000)),
    (1.000, (0.000, 0.467, 0.988)))) 

sns.set("paper", "darkgrid")
plt.figure(dpi=300,figsize=(15,15))
with open("heat_map_vehicle_and_alert.pkl", "rb+") as f:
    heat_map = pickle.load(f)
# sns.set(rc={'figure.figsize': (10.5, 80)})
# center = -60
heat_map = np.array(heat_map)
heat_map = (np.array(heat_map)+140)*6
heat_map = np.clip(heat_map,-400, 200)

ax = sns.heatmap(np.array(heat_map)[:,:], linewidth=0, cmap=my_color, robust=True)
ax.set_aspect("equal")
ax.set_xticks([])
ax.set_yticks([])
```
